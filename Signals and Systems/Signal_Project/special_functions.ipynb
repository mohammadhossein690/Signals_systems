```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(-2, 2, 1000) 
u = np.where(t >= 0, 1, 0)
plt.figure(figsize=(8, 4))
plt.plot(t, u, label='u(t)', color='blue')
plt.title('Unit Step Signal')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axvline(0, color='black', lw=1)  
plt.axhline(0, color='black', lw=1)
plt.ylim(-0.1, 1.1)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def unit_doublet(t, epsilon=1e-3):
    return np.where(np.abs(t) < epsilon, 2 / epsilon**2, 0)
t = np.linspace(-1, 1, 1000)
y = unit_doublet(t)
plt.plot(t, y)
plt.title('Unit Doublet Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 10, 1000)  
signal_ramp = t  
plt.figure(figsize=(10, 5))
plt.plot(t, signal_ramp, label='Ramp Signal', color='blue')
plt.title('Continuous Ramp Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
fs = 1000   
f = 5      
tau = 1  
t = np.linspace(0, 2, fs * 2) 
signal = np.sin(2 * np.pi * f * t) * np.exp(-t / tau)
plt.figure(figsize=(10, 6))
plt.plot(t, signal)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(-5, 6)  
x = np.zeros_like(n)  
x[n == 0] = 1 
plt.stem(n, x, use_line_collection=True)
plt.ylabel('amplitude')
plt.grid()
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.xlim(-5, 5)  
plt.ylim(-0.5, 1.5) 
plt.show()
```
