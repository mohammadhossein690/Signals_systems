```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 20 * np.pi, 1000) 
x_t = np.cos(t / 6)
plt.figure(figsize=(10, 4))
plt.plot(t, x_t, label=r'$x(t) = \cos\left(\frac{1}{6}t\right)$')
plt.xlabel('Time (t)')
plt.ylabel('x(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 20, 1000)
x_t = np.cos((8 * np.pi / 31) * t)
plt.figure(figsize=(8, 4))
plt.plot(t, x_t, label=r'$x(t) = \cos\left(\frac{8\pi}{31}t\right)$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(0, 100)
x_n = np.cos((8 * np.pi / 31) * n)
plt.figure(figsize=(10, 4))
plt.stem(n, x_n, use_line_collection=True)
plt.title('Signal x[n] = cos(8π/31 * n)')
plt.xlabel('n')
plt.ylabel('x[n]')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 2 * np.pi, 1000)
x_t = np.exp(1j * 2 * t) + np.exp(1j * 3 * t)
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, np.real(x_t), label='Real Part')
plt.title('Real Part of x(t) = e^{j2t} + e^{j3t}')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, np.imag(x_t), label='Imaginary Part')
plt.title('Imaginary Part of x(t) = e^{j2t} + e^{j3t}')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(0, 100, 3) 
x_t = np.exp(1j * (2 * np.pi / 3) * n) + np.exp(1j * (3 * np.pi / 4) * n)
plt.figure(figsize=(14, 6))
plt.subplot(2, 1, 1)
plt.plot(n, np.real(x_t), label='Real Part')
plt.title('Real Part of x(t)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(n, np.imag(x_t), label='Imaginary Part', color='orange')
plt.title('Imaginary Part of x(t)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0, 500, 1000)  
x_t = 6 * np.exp(3 * t) * np.cos((1/6) * t)

plt.figure(figsize=(10, 5))
plt.plot(t, x_t, label=r'$x(t) = 6e^{3t} \cos\left(\frac{1}{6}t\right)$')
plt.title('Plot of the Signal x(t)')
plt.xlabel('Time (t)')
plt.ylabel('x(t)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.ylim(-10, 10) 
plt.show()
```
